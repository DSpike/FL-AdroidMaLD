```python
import pandas as pd
import numpy as np
import glob
import os
from sklearn.model_selection import train_test_split, KFold
from sklearn.preprocessing import LabelEncoder
from sklearn.impute import SimpleImputer
from sklearn.metrics import classification_report, confusion_matrix, accuracy_score
import matplotlib.pyplot as plt
import warnings
import itertools
import xgboost as xgb
import time
import torch
import torch.nn as nn
import torch.optim as optim
from torch.utils.data import DataLoader, TensorDataset
import math

# Ensure results directory exists
results_dir = r'C:\Users\USER\Documents\NTUST\Conference_Workshop_Seminar\FL-AdroidMaLD\results'
os.makedirs(results_dir, exist_ok=True)

# Suppress warnings
warnings.filterwarnings('ignore')

# MGU Implementation
class MGU(nn.Module):
    def __init__(self, input_size, hidden_size, batch_first=True):
        super(MGU, self).__init__()
        self.input_size = input_size
        self.hidden_size = hidden_size
        self.batch_first = batch_first

        # Forget gate
        self.forget = nn.Linear(input_size + hidden_size, hidden_size)
        # Candidate hidden state
        self.hidden = nn.Linear(input_size + hidden_size, hidden_size)

        self.sigmoid = nn.Sigmoid()
        self.tanh = nn.Tanh()

    def forward(self, x, h=None):
        if self.batch_first:
            batch_size, seq_len, _ = x.size()
        else:
            seq_len, batch_size, _ = x.size()

        if h is None:
            h = torch.zeros(batch_size, self.hidden_size).to(x.device)

        outputs = []
        for t in range(seq_len):
            x_t = x[:, t, :] if self.batch_first else x[t, :, :]
            combined = torch.cat((x_t, h), dim=1)

            # Forget gate
            f_t = self.sigmoid(self.forget(combined))
            # Candidate hidden state
            h_tilde = self.tanh(self.hidden(combined))
            # New hidden state
            h = (1 - f_t) * h + f_t * h_tilde
            outputs.append(h.unsqueeze(1 if self.batch_first else 0))

        outputs = torch.cat(outputs, dim=1 if self.batch_first else 0)
        return outputs, h

# Original CNN-MGU model with standard convolution
class CNNMGU(nn.Module):
    def __init__(self, input_dim, hidden_dim, num_classes):
        super(CNNMGU, self).__init__()
        self.conv1 = nn.Conv1d(1, 16, kernel_size=3, padding=1)
        self.relu = nn.ReLU()
        self.pool = nn.MaxPool1d(2)
        self.mgu = MGU(input_size=16, hidden_size=hidden_dim, batch_first=True)
        self.fc = nn.Linear(hidden_dim, num_classes)

    def forward(self, x):
        x = x.unsqueeze(1)  # (batch_size, 1, input_dim)
        x = self.conv1(x)   # (batch_size, 16, input_dim)
        x = self.relu(x)
        x = self.pool(x)    # (batch_size, 16, input_dim // 2)
        x = x.permute(0, 2, 1)  # (batch_size, input_dim // 2, 16)
        x, hn = self.mgu(x)  # (batch_size, input_dim // 2, hidden_dim)
        x = self.fc(hn)      # (batch_size, num_classes)
        return x

# CNN-MGU model with Depthwise Separable Convolution
class CNNMGU_Depthwise(nn.Module):
    def __init__(self, input_dim, hidden_dim, num_classes):
        super(CNNMGU_Depthwise, self).__init__()
        self.depthwise = nn.Conv1d(1, 1, kernel_size=3, padding=1, groups=1)
        self.pointwise = nn.Conv1d(1, 16, kernel_size=1)
        self.relu = nn.ReLU()
        self.pool = nn.MaxPool1d(2)
        self.mgu = MGU(input_size=16, hidden_size=hidden_dim, batch_first=True)
        self.fc = nn.Linear(hidden_dim, num_classes)

    def forward(self, x):
        x = x.unsqueeze(1)  # (batch_size, 1, input_dim)
        x = self.depthwise(x)  # (batch_size, 1, input_dim)
        x = self.pointwise(x)  # (batch_size, 16, input_dim)
        x = self.relu(x)
        x = self.pool(x)    # (batch_size, 16, input_dim // 2)
        x = x.permute(0, 2, 1)  # (batch_size, input_dim // 2, 16)
        x, hn = self.mgu(x)  # (batch_size, input_dim // 2, hidden_dim)
        x = self.fc(hn)      # (batch_size, num_classes)
        return x

# Function to initialize server-side weights for both models
def initialize_server_weights(n_features, n_classes):
    hidden_dim = 64
    initial_weights = {}
    
    # Initialize CNNMGU_Depthwise weights
    model_depthwise = CNNMGU_Depthwise(n_features, hidden_dim, n_classes)
    initial_weights['cnn_mgu_depthwise'] = {key: value.clone() for key, value in model_depthwise.state_dict().items()}
    
    # Initialize CNNMGU weights
    model_original = CNNMGU(n_features, hidden_dim, n_classes)
    initial_weights['cnn_mgu_original'] = {key: value.clone() for key, value in model_original.state_dict().items()}
    
    return initial_weights

# Function to load and preprocess the dataset
def load_dataset(filename):
    print(f"Loading dataset: {filename}")
    df = pd.read_csv(filename)
    if "Category" not in df.columns:
        raise ValueError("'Category' column not found in the dataset")
    features = [col for col in df.columns if col not in ["Category", "Family"]]
    print(f"Original number of features: {len(features)}")
    for col in features:
        if df[col].dtype == 'object':
            le = LabelEncoder()
            df[col] = le.fit_transform(df[col])
    imputer = SimpleImputer(strategy='mean')
    df[features] = imputer.fit_transform(df[features])
    X = df[features].values
    le = LabelEncoder()
    y = le.fit_transform(df["Category"])
    class_names = le.classes_
    class_counts = pd.Series(df["Category"]).value_counts()
    print(f"Target classes: {class_names}")
    print("Class distribution:")
    for cls, count in class_counts.items():
        print(f"  {cls}: {count}")
    return X, y, features, class_names

# Function to plot sample size of each class before distribution
def plot_class_sample_size(y, class_names):
    unique_classes, counts = np.unique(y, return_counts=True)
    plt.figure(figsize=(10, 6))
    plt.bar([class_names[cls] for cls in unique_classes], counts, color='skyblue')
    plt.title('Sample Size of Each Class Before Distribution', fontsize=14)
    plt.xlabel('Classes', fontsize=12)
    plt.ylabel('Number of Samples', fontsize=12)
    plt.xticks(rotation=45, ha='right')
    for i, count in enumerate(counts):
        plt.text(i, count + 5, str(count), ha='center', va='bottom')
    plt.tight_layout()
    plt.savefig(os.path.join(results_dir, 'class_sample_size_before_distribution.png'))
    plt.show()

# Function to split data for federated learning clients and visualize distribution
def create_federated_clients(X, y, n_clients=5, strategy='iid', class_names=None):
    clients_data = []
    unique_classes = np.unique(y)
    client_class_distributions = []
    plot_class_sample_size(y, class_names)
    if strategy == 'iid':
        indices = np.random.permutation(len(X))
        client_size = len(indices) // n_clients
        for i in range(n_clients):
            start_idx = i * client_size
            end_idx = (i + 1) * client_size if i < n_clients - 1 else len(indices)
            client_indices = indices[start_idx:end_idx]
            clients_data.append((X[client_indices], y[client_indices]))
            class_dist = np.unique(y[client_indices], return_counts=True)
            client_class_distributions.append(dict(zip(class_dist[0], class_dist[1])))
    elif strategy == 'non_iid':
        n_classes = len(unique_classes)
        alpha = 3
        class_indices = {cls: np.where(y == cls)[0] for cls in unique_classes}
        client_distributions = np.random.dirichlet([alpha] * n_classes, n_clients)
        for i in range(n_clients):
            client_indices = []
            client_distribution = client_distributions[i]
            for cls_idx, cls in enumerate(unique_classes):
                target_proportion = client_distribution[cls_idx]
                available_indices = class_indices[cls]
                target_count = int(target_proportion * len(available_indices))
                target_count = min(target_count, len(available_indices))
                if target_count > 0:
                    sampled_indices = np.random.choice(available_indices, size=target_count, replace=False)
                    client_indices.extend(sampled_indices)
            client_indices = list(set(client_indices))
            np.random.shuffle(client_indices)
            if len(np.unique(y[client_indices])) < 2:
                print(f"Skipping Client {i + 1} due to insufficient class representation.")
                continue
            clients_data.append((X[client_indices], y[client_indices]))
            class_dist = np.unique(y[client_indices], return_counts=True)
            client_class_distributions.append(dict(zip(class_dist[0], class_dist[1])))
            print(f"Client {i + 1} assigned {len(client_indices)} samples with classes: {np.unique(y[client_indices], return_counts=True)}")
    plot_client_distribution(client_class_distributions, unique_classes, n_clients, class_names)
    return clients_data

# Function to plot bar graph for client class and sample distribution
def plot_client_distribution(client_class_distributions, unique_classes, n_clients, class_names):
    plt.figure(figsize=(12, 6))
    clients = [f"Client {i+1}" for i in range(n_clients)]
    class_counts_per_client = np.zeros((n_clients, len(unique_classes)))
    for i, dist in enumerate(client_class_distributions):
        for cls, count in dist.items():
            class_counts_per_client[i, cls] = count
    bottom = np.zeros(n_clients)
    for idx, cls in enumerate(unique_classes):
        counts = class_counts_per_client[:, cls]
        plt.bar(clients, counts, bottom=bottom, label=f'{class_names[cls]}')
        bottom += counts
    plt.title('Class and Sample Distribution Across Clients', fontsize=14)
    plt.xlabel('Clients', fontsize=12)
    plt.ylabel('Number of Samples', fontsize=12)
    plt.legend(title='Classes', bbox_to_anchor=(1.05, 1), loc='upper left')
    total_samples = class_counts_per_client.sum(axis=1)
    for i, total in enumerate(total_samples):
        plt.text(i, total + 5, int(total), ha='center', va='bottom')
    plt.tight_layout()
    plt.savefig(os.path.join(results_dir, 'client_class_distribution.png'))
    plt.show()

# Evaluate features using CNN-MGU with global n_classes
def evaluate_features(X, y, features_idx, n_classes_global, model_class=CNNMGU):
    start_time = time.time()
    if len(features_idx) == 0:
        return 0, 0
    selected_features = X[:, features_idx.astype(bool)]
    if selected_features.shape[1] == 0:
        return 0, 0
    unique, counts = np.unique(y, return_counts=True)
    class_counts = dict(zip(unique, counts))
    for cls, count in class_counts.items():
        if count < 2:
            print(f"Class {cls} has only {count} sample(s), which is too few. Removing this class.")
            mask = y != cls
            selected_features = selected_features[mask]
            y = y[mask]
    X_train, X_test, y_train, y_test = train_test_split(
        selected_features, y, test_size=0.3, random_state=42, stratify=y
    )
    X_train_tensor = torch.FloatTensor(X_train)
    y_train_tensor = torch.LongTensor(y_train)
    X_test_tensor = torch.FloatTensor(X_test)
    y_test_tensor = torch.LongTensor(y_test)
    train_dataset = TensorDataset(X_train_tensor, y_train_tensor)
    train_loader = DataLoader(train_dataset, batch_size=32, shuffle=True)
    hidden_dim = 64
    cnn_mgu_model = model_class(X_train.shape[1], hidden_dim, n_classes_global)
    optimizer = optim.Adam(cnn_mgu_model.parameters(), lr=0.001)
    criterion = nn.CrossEntropyLoss()
    cnn_mgu_model.train()
    for epoch in range(15):
        for data, target in train_loader:
            optimizer.zero_grad()
            output = cnn_mgu_model(data)
            loss = criterion(output, target)
            loss.backward()
            optimizer.step()
    cnn_mgu_model.eval()
    with torch.no_grad():
        output = cnn_mgu_model(X_test_tensor)
        _, predicted = torch.max(output, 1)
        accuracy = (predicted == y_test_tensor).sum().item() / len(y_test_tensor)
    comp_time = time.time() - start_time
    return accuracy, comp_time

# XGBoost Feature Selection
def xgboost_feature_selection(X, y, n_features, n_classes_global):
    start_time = time.time()
    n_classes_local = len(np.unique(y))
    xgb_model = xgb.XGBClassifier(
        objective='multi:softmax',
        num_class=n_classes_local,
        use_label_encoder=False,
        eval_metric='mlogloss',
        n_jobs=-1
    )
    xgb_model.fit(X, y)
    feature_importance = xgb_model.feature_importances_
    sorted_idx = np.argsort(feature_importance)[::-1]
    selected_features_idx = np.zeros(n_features, dtype=bool)
    top_n = min(int(n_features * 0.5), n_features)
    selected_features_idx[sorted_idx[:top_n]] = True
    total_time = time.time() - start_time
    accuracy, eval_time = evaluate_features(X, y, selected_features_idx, n_classes_global)
    return selected_features_idx, total_time, eval_time

# Train client model with both CNN-MGU versions
def train_client_model(client_id, X, y, features, class_names, server_weights, n_classes_global, round_num):
    print(f"\n===== Training Client {client_id} (Round {round_num}) =====")
    xgb_time_total = 0
    xgb_eval_time_total = 0
    start_xgb = time.time()
    selected_features_idx, xgb_time, xgb_eval_time = xgboost_feature_selection(X, y, X.shape[1], n_classes_global)
    xgb_time_total = time.time() - start_xgb
    xgb_eval_time_total = xgb_eval_time
    X_selected = X[:, selected_features_idx.astype(bool)]
    X_tensor = torch.FloatTensor(X_selected)
    y_tensor = torch.LongTensor(y)
    dataset = TensorDataset(X_tensor, y_tensor)
    train_loader = DataLoader(dataset, batch_size=32, shuffle=True)
    hidden_dim = 64
    
    # Original CNN-MGU
    cnn_mgu_original = CNNMGU(X_selected.shape[1], hidden_dim, n_classes_global)
    cnn_mgu_original.load_state_dict(server_weights['cnn_mgu_original'])
    optimizer_orig = optim.Adam(cnn_mgu_original.parameters(), lr=0.001)
    criterion = nn.CrossEntropyLoss()

    # Depthwise CNN-MGU
    cnn_mgu_depthwise = CNNMGU_Depthwise(X_selected.shape[1], hidden_dim, n_classes_global)
    cnn_mgu_depthwise.load_state_dict(server_weights['cnn_mgu_depthwise'])
    optimizer_depth = optim.Adam(cnn_mgu_depthwise.parameters(), lr=0.001)

    kf = KFold(n_splits=5, shuffle=True, random_state=42)
    train_accuracies_orig = []
    val_accuracies_orig = []
    train_accuracies_depth = []
    val_accuracies_depth = []
    training_time_orig = 0
    training_time_depth = 0

    for fold, (train_idx, val_idx) in enumerate(kf.split(X_selected)):
        X_train, X_val = X_selected[train_idx], X_selected[val_idx]
        y_train, y_val = y[train_idx], y[val_idx]
        X_train_tensor = torch.FloatTensor(X_train)
        y_train_tensor = torch.LongTensor(y_train)
        X_val_tensor = torch.FloatTensor(X_val)
        y_val_tensor = torch.LongTensor(y_val)
        train_dataset = TensorDataset(X_train_tensor, y_train_tensor)
        train_loader = DataLoader(train_dataset, batch_size=32, shuffle=True)

        fold_train_acc_orig = []
        fold_val_acc_orig = []
        fold_train_acc_depth = []
        fold_val_acc_depth = []

        # Train Original Model
        start_time_orig = time.time()
        for epoch in range(10):
            cnn_mgu_original.train()
            for data, target in train_loader:
                optimizer_orig.zero_grad()
                output = cnn_mgu_original(data)
                loss = criterion(output, target)
                loss.backward()
                optimizer_orig.step()
            cnn_mgu_original.eval()
            with torch.no_grad():
                train_output = cnn_mgu_original(X_train_tensor)
                _, train_pred = torch.max(train_output, 1)
                train_acc = (train_pred == y_train_tensor).sum().item() / len(y_train_tensor)
                fold_train_acc_orig.append(train_acc)
                val_output = cnn_mgu_original(X_val_tensor)
                _, val_pred = torch.max(val_output, 1)
                val_acc = (val_pred == y_val_tensor).sum().item() / len(y_val_tensor)
                fold_val_acc_orig.append(val_acc)
        training_time_orig += time.time() - start_time_orig

        # Train Depthwise Model
        start_time_depth = time.time()
        for epoch in range(10):
            cnn_mgu_depthwise.train()
            for data, target in train_loader:
                optimizer_depth.zero_grad()
                output = cnn_mgu_depthwise(data)
                loss = criterion(output, target)
                loss.backward()
                optimizer_depth.step()
            cnn_mgu_depthwise.eval()
            with torch.no_grad():
                train_output = cnn_mgu_depthwise(X_train_tensor)
                _, train_pred = torch.max(train_output, 1)
                train_acc = (train_pred == y_train_tensor).sum().item() / len(y_train_tensor)
                fold_train_acc_depth.append(train_acc)
                val_output = cnn_mgu_depthwise(X_val_tensor)
                _, val_pred = torch.max(val_output, 1)
                val_acc = (val_pred == y_val_tensor).sum().item() / len(y_val_tensor)
                fold_val_acc_depth.append(val_acc)
        training_time_depth += time.time() - start_time_depth

        train_accuracies_orig.append(fold_train_acc_orig)
        val_accuracies_orig.append(fold_val_acc_orig)
        train_accuracies_depth.append(fold_train_acc_depth)
        val_accuracies_depth.append(fold_val_acc_depth)

    train_accuracies_orig = np.mean(train_accuracies_orig, axis=0)
    val_accuracies_orig = np.mean(val_accuracies_orig, axis=0)
    train_accuracies_depth = np.mean(train_accuracies_depth, axis=0)
    val_accuracies_depth = np.mean(val_accuracies_depth, axis=0)
    
    accuracy_orig = val_accuracies_orig[-1]
    accuracy_depth = val_accuracies_depth[-1]
    selected_features_names = [features[i] for i in range(len(selected_features_idx)) if selected_features_idx[i]]

    X_train_full, X_test, y_train_full, y_test = train_test_split(X_selected, y, test_size=0.3, random_state=42)
    X_test_tensor = torch.FloatTensor(X_test)
    y_test_tensor = torch.LongTensor(y_test)
    
    cnn_mgu_original.eval()
    with torch.no_grad():
        test_output_orig = cnn_mgu_original(X_test_tensor)
        _, test_pred_orig = torch.max(test_output_orig, 1)
        test_accuracy_orig = (test_pred_orig == y_test_tensor).sum().item() / len(y_test_tensor)
    
    cnn_mgu_depthwise.eval()
    with torch.no_grad():
        test_output_depth = cnn_mgu_depthwise(X_test_tensor)
        _, test_pred_depth = torch.max(test_output_depth, 1)
        test_accuracy_depth = (test_pred_depth == y_test_tensor).sum().item() / len(y_test_tensor)

    total_params_orig = sum(p.numel() for p in cnn_mgu_original.parameters())
    total_params_depth = sum(p.numel() for p in cnn_mgu_depthwise.parameters())

    client_results = {
        'client_id': client_id,
        'classifier_orig': cnn_mgu_original,
        'classifier_depth': cnn_mgu_depthwise,
        'accuracy_orig': accuracy_orig,
        'accuracy_depth': accuracy_depth,
        'test_accuracy_orig': test_accuracy_orig,
        'test_accuracy_depth': test_accuracy_depth,
        'selected_features_idx': selected_features_idx,
        'selected_features_names': selected_features_names,
        'class_names': class_names,
        'train_accuracies_orig': train_accuracies_orig.tolist(),
        'val_accuracies_orig': val_accuracies_orig.tolist(),
        'train_accuracies_depth': train_accuracies_depth.tolist(),
        'val_accuracies_depth': val_accuracies_depth.tolist(),
        'xgb_time': xgb_time_total,
        'xgb_eval_time': xgb_eval_time_total,
        'training_time_orig': training_time_orig,
        'training_time_depth': training_time_depth,
        'total_params_orig': total_params_orig,
        'total_params_depth': total_params_depth
    }

    print(f"Client {client_id} - Original CNN-MGU:")
    print(f"  Final accuracy (5-fold CV avg): {accuracy_orig * 100:.2f}%")
    print(f"  Local test accuracy: {test_accuracy_orig * 100:.2f}%")
    print(f"  Training time: {training_time_orig:.2f}s")
    print(f"  Total parameters: {total_params_orig}")
    
    print(f"Client {client_id} - Depthwise CNN-MGU:")
    print(f"  Final accuracy (5-fold CV avg): {accuracy_depth * 100:.2f}%")
    print(f"  Local test accuracy: {test_accuracy_depth * 100:.2f}%")
    print(f"  Training time: {training_time_depth:.2f}s")
    print(f"  Total parameters: {total_params_depth}")
    
    print(f"Comparison:")
    print(f"  Parameter reduction: {total_params_orig - total_params_depth} ({(total_params_orig - total_params_depth) / total_params_orig * 100:.2f}%)")
    print(f"  Training time reduction: {training_time_orig - training_time_depth:.2f}s ({(training_time_orig - training_time_depth) / training_time_orig * 100:.2f}%)")
    
    print(f"Selected features: {selected_features_names}")
    print(f"Number of selected features: {len(selected_features_names)} out of {len(features)} ({len(selected_features_names)/len(features)*100:.2f}%)")
    print(f"XGBoost time: {xgb_time_total:.2f}s")
    print(f"XGBoost eval time: {xgb_eval_time_total:.2f}s")

    plot_accuracy(train_accuracies_depth, val_accuracies_depth, client_id, model_name="Depthwise CNN-MGU")
    plot_accuracy(train_accuracies_orig, val_accuracies_orig, client_id, model_name="Original CNN-MGU")

    return client_results

# Function to plot training and validation accuracy
def plot_accuracy(train_acc, val_acc, client_id, model_name):
    plt.figure(figsize=(8, 6))
    plt.plot(train_acc, label='Training Accuracy', color='blue', linewidth=2)
    plt.plot(val_acc, label='Validation Accuracy', color='red', linewidth=2)
    plt.title(f'Client {client_id} - {model_name} Performance (5-Fold CV)', fontsize=12, pad=10)
    plt.xlabel('Epochs', fontsize=10)
    plt.ylabel('Accuracy', fontsize=10)
    plt.legend(loc='upper right', fontsize=10)
    plt.ylim(0.7, 1.0)
    plt.grid(False)
    plt.tight_layout()
    plt.savefig(os.path.join(results_dir, f'client_{client_id}_accuracy_{model_name.replace(" ", "_").lower()}.png'))
    plt.show()

# Federated aggregation for both models
def federated_aggregation_secure(client_results, n_features, n_classes_global):
    print("\n===== Federated Model Aggregation (Secure) =====")
    n_clients = len(client_results)
    hidden_dim = 64
    
    global_model_depth = CNNMGU_Depthwise(n_features, hidden_dim, n_classes_global)
    global_state_dict_depth = global_model_depth.state_dict()
    for key in global_state_dict_depth.keys():
        global_state_dict_depth[key] = torch.zeros_like(global_state_dict_depth[key])
    
    global_model_orig = CNNMGU(n_features, hidden_dim, n_classes_global)
    global_state_dict_orig = global_model_orig.state_dict()
    for key in global_state_dict_orig.keys():
        global_state_dict_orig[key] = torch.zeros_like(global_state_dict_orig[key])

    weight_count = 0
    for client in client_results:
        client_model_depth = client['classifier_depth']
        client_state_dict_depth = client_model_depth.state_dict()
        for key in global_state_dict_depth.keys():
            global_state_dict_depth[key] += client_state_dict_depth[key]
        
        client_model_orig = client['classifier_orig']
        client_state_dict_orig = client_model_orig.state_dict()
        for key in global_state_dict_orig.keys():
            global_state_dict_orig[key] += client_state_dict_orig[key]
        
        weight_count += 1

    if weight_count > 0:
        for key in global_state_dict_depth.keys():
            global_state_dict_depth[key] /= weight_count
        for key in global_state_dict_orig.keys():
            global_state_dict_orig[key] /= weight_count
        print(f"Aggregated weights from {weight_count} clients for both models.")
    
    global_model_depth.load_state_dict(global_state_dict_depth)
    global_model_orig.load_state_dict(global_state_dict_orig)

    client_test_accuracies_depth = [client['test_accuracy_depth'] for client in client_results]
    client_test_accuracies_orig = [client['test_accuracy_orig'] for client in client_results]
    avg_test_accuracy_depth = np.mean(client_test_accuracies_depth)
    avg_test_accuracy_orig = np.mean(client_test_accuracies_orig)

    print(f"Average client-reported test accuracy (Depthwise): {avg_test_accuracy_depth * 100:.2f}%")
    print(f"Average client-reported test accuracy (Original): {avg_test_accuracy_orig * 100:.2f}%")

    return {
        'accuracy_depth': avg_test_accuracy_depth,
        'accuracy_orig': avg_test_accuracy_orig,
        'classifier_depth': global_model_depth,
        'classifier_orig': global_model_orig,
        'server_weights': {
            'cnn_mgu_depthwise': global_state_dict_depth,
            'cnn_mgu_original': global_state_dict_orig
        }
    }

# Compare models using client test accuracies and federated average
def compare_models(client_results, federated_results):
    print("\n===== Model Performance Comparison =====")
    models = ['Federated Depthwise', 'Federated Original']
    accuracies = [federated_results['accuracy_depth'] * 100, federated_results['accuracy_orig'] * 100]
    for client in client_results:
        models.append(f"Client {client['client_id']} (Depth)")
        models.append(f"Client {client['client_id']} (Orig)")
        accuracies.append(client['test_accuracy_depth'] * 100)
        accuracies.append(client['test_accuracy_orig'] * 100)
    fig, ax = plt.subplots(figsize=(12, 6))
    ax.bar(models, accuracies, color=['red', 'green'] + ['blue', 'orange'] * len(client_results))
    ax.set_title('Model Accuracy Comparison (Client Test Accuracies vs. Federated Avg)')
    ax.set_ylabel('Accuracy (%)')
    ax.set_ylim([min(accuracies) - 5, max(accuracies) + 5])
    for i, v in enumerate(accuracies):
        ax.text(i, v + 0.5, f'{v:.2f}%', ha='center', va='bottom')
    plt.xticks(rotation=45, ha='right')
    plt.tight_layout()
    plt.savefig(os.path.join(results_dir, 'model_comparison.png'))
    plt.show()

# Main function
def main():
    data_path = r'C:\Users\USER\Documents\NTUST\Conference_Workshop_Seminar\FL-AdroidMaLD\normalized_dataset'
    os.makedirs(data_path, exist_ok=True)
    csv_files = glob.glob(os.path.join(data_path, "*after_reboot*.csv"))
    if not csv_files:
        print(f"No 'after_reboot' CSV files found in the directory: {data_path}")
        return
    combined_csv_file = os.path.join(data_path, "combined_dataset.csv")
    if not os.path.exists(combined_csv_file):
        combined_df = pd.DataFrame()
        for csv_file in csv_files:
            df = pd.read_csv(csv_file)
            combined_df = pd.concat([combined_df, df], ignore_index=True)
        combined_df.to_csv(combined_csv_file, index=False)
    X, y, features, class_names = load_dataset(combined_csv_file)
    print(f"Global unique classes: {np.unique(y)}")
    print(f"n_classes_global: {len(np.unique(y))}")
    n_classes_global = len(np.unique(y))
    server_weights = initialize_server_weights(n_features=X.shape[1], n_classes=n_classes_global)
    n_clients = 5
    clients_data = create_federated_clients(X, y, n_clients=n_clients, strategy='non_iid', class_names=class_names)
    for i, (client_X, client_y) in enumerate(clients_data):
        print(f"Client {i+1} unique classes: {np.unique(client_y)}")
    n_rounds = 9
    all_client_results = []
    federated_results = None
    n_clients_selected = 3
    for round_num in range(1, n_rounds + 1):
        print(f"\n===== Communication Round {round_num} =====")
        if round_num == 1:
            client_results = []
            for i, (client_X, client_y) in enumerate(clients_data):
                print(f"\nClient {i + 1} has {len(client_X)} samples.")
                client_result = train_client_model(i + 1, client_X, client_y, features, class_names, server_weights, n_classes_global, round_num)
                client_results.append(client_result)
        else:
            client_metrics = all_client_results[-1]
            selected_indices = aquila_optimization_client_selection(client_metrics, n_clients, n_clients_selected)
            client_results = []
            for idx in selected_indices:
                client_X, client_y = clients_data[idx]
                print(f"\nTraining selected Client {idx + 1} with {len(client_X)} samples.")
                client_result = train_client_model(idx + 1, client_X, client_y, features, class_names, server_weights, n_classes_global, round_num)
                client_results.append(client_result)
        all_client_results.append(client_results)
        federated_results = federated_aggregation_secure(client_results, len(features), n_classes_global)
        server_weights = federated_results['server_weights']
        compare_models(client_results, federated_results)

def aquila_optimization_client_selection(client_metrics, n_clients_total, n_clients_selected, max_iter=20):
    n_available_clients = len(client_metrics)
    if n_clients_selected > n_available_clients:
        n_clients_selected = n_available_clients
    pop_size = 10
    population = np.zeros((pop_size, n_available_clients))
    for i in range(pop_size):
        indices = np.random.choice(n_available_clients, n_clients_selected, replace=False)
        population[i, indices] = 1
    def fitness(solution):
        selected_indices = np.where(solution == 1)[0]
        if len(selected_indices) == 0:
            return -np.inf
        accuracies = [client_metrics[i]['test_accuracy_depth'] for i in selected_indices]
        comp_times = [client_metrics[i]['xgb_time'] for i in selected_indices]
        avg_accuracy = np.mean(accuracies)
        avg_comp_time = np.mean(comp_times)
        return 0.8 * avg_accuracy - 0.2 * avg_comp_time
    def levy_flight(d):
        sigma = (math.gamma(1 + 1.5) * np.sin(np.pi * 1.5 / 2) / 
                 (math.gamma((1 + 1.5) / 2) * 1.5 * 2 ** ((1.5 - 1) / 2))) ** (1 / 1.5)
        u = np.random.normal(0, sigma, d)
        v = np.random.normal(0, 1, d)
        return 0.01 * u / (np.abs(v) ** (1 / 1.5))
    best_solution = population[0].copy()
    best_fitness = fitness(best_solution)
    for t in range(1, max_iter + 1):
        X_mean = np.mean(population, axis=0)
        for i in range(pop_size):
            r1, r2, r3, r4, r5, r6 = np.random.rand(6)
            if t <= (2/3) * max_iter:
                if r1 < 0.5:
                    X_rand = population[np.random.randint(pop_size)]
                    population[i] = best_solution * (1 - t/max_iter) + (X_rand - best_solution) * r2
                else:
                    population[i] = best_solution * levy_flight(n_available_clients) + X_mean + r3 * (1 - 0) * r4
            else:
                if r1 < 0.5:
                    population[i] = (best_solution - population[i]) * r2 - r3 * (1 - 0) * (t/max_iter)
                else:
                    G1 = 2 * r4 - 1
                    G2 = 2 * (1 - t/max_iter)
                    population[i] = best_solution - G1 * (r5 * (1 - 0)) + G2 * levy_flight(n_available_clients)
            population[i] = np.clip(population[i], 0, 1)
            sorted_indices = np.argsort(population[i])[::-1]
            population[i] = np.zeros(n_available_clients)
            population[i, sorted_indices[:n_clients_selected]] = 1
            curr_fitness = fitness(population[i])
            if curr_fitness > best_fitness:
                best_fitness = curr_fitness
                best_solution = population[i].copy()
    selected_indices = np.where(best_solution == 1)[0]
    original_client_ids = [client_metrics[i]['client_id'] - 1 for i in selected_indices]
    print(f"AOA selected clients (original IDs): {np.array(original_client_ids) + 1}")
    return original_client_ids

if __name__ == "__main__":
    start_time = time.time()
    main()
    print(f"\nTotal execution time: {time.time() - start_time:.2f} seconds")
```
